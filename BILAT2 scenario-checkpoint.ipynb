```python
# Loading data
data_table =  pd.read_excel("./data/DATABASE.xls", header=3, nrows=204)

# Taking a look at head
data_table.head()
```

```python
# Removing last characters of column names (removing exponents)
cols = list(data_table.columns)

for i in range(39):
    if(i > 25):
        cols[i] = cols[i][:-2]
    
    else:
        cols[i] = cols[i][:-1]
        
# Replacing column names of the dataframe
col_rename_dict = {i:j for i,j in zip(list(data_table.columns), cols)}
data_table.rename(columns=col_rename_dict, inplace=True)

# Handling exception renaming
data_table.rename(columns={"Goa":"Goal"})
```
